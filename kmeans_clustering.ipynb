```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("student_clustering.csv")
```

```python
df.shape
```

```python
plt.scatter(df["cgpa"],df["iq"])
```

```python
from sklearn.cluster import KMeans
```

```python
wcss = []
for i in range(1,11):
    km = KMeans(n_clusters=i)
    km.fit_predict(df)
    wcss.append(km.inertia_)
wcss
```

```python
plt.plot(range(1,11),wcss)
```

```python
#we can see that the 4 is the elbow point 
X=df.iloc[:,:].values
km_final=KMeans(n_clusters=4)
y_pred=km_final.fit_predict(X)
```

```python
y_pred
```

```python
plt.scatter(X[y_pred==0,0],X[y_pred==0,1],color="red")
plt.scatter(X[y_pred==1,0],X[y_pred==1,1],color="green")
plt.scatter(X[y_pred==2,0],X[y_pred==2,1],color="black")
plt.scatter(X[y_pred==3,0],X[y_pred==3,1],color="pink")
```
